```python
import numpy as np

def gradient_descent(X, Y, W, alpha, iterations):
    """
    Perform gradient descent to optimize linear regression parameters.

    X: Feature matrix (d Ã— n)
    Y: Target vector (n,)
    W: Weight vector (d,)
    """
    n = Y.shape[0]   # number of samples
    cost_history = []

    for i in range(iterations):
        # Prediction
        Y_pred = np.dot(W.T, X)   # (n,)

        # Gradient computation
        gradient = (1 / n) * np.dot(X, (Y_pred - Y).T)  # (d,)

        # Weight update
        W = W - alpha * gradient

        # Cost (MSE)
        cost = (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)
        cost_history.append(cost)

    return W, cost_history


# ===== Test with given example =====
X = np.array([[1, 3, 5],
              [2, 4, 6]])   # (d=2, n=3)

Y = np.array([3, 7, 11])
W_init = np.zeros(X.shape[0])

alpha = 0.01
iterations = 1000

W_final, cost_hist = gradient_descent(X, Y, W_init, alpha, iterations)

print("Updated weights:", W_final)
print("Final cost:", cost_hist[-1])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
X = np.array([0, 1, 2, 3, 4, 5])
Y_actual = np.array([2.5, 3.6, 6.1, 7.5, 8.8, 10.1])  # actual values
Y_pred = np.array([2.7, 3.5, 6.0, 7.4, 8.9, 10.0])    # predicted values

rmse_val = np.sqrt(np.mean((Y_actual - Y_pred) ** 2))

ssr = np.sum((Y_actual - Y_pred) ** 2)
sst = np.sum((Y_actual - np.mean(Y_actual)) ** 2)
r2_val = 1 - (ssr / sst)

plt.figure(figsize=(8,6))
plt.scatter(X, Y_actual, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Predicted Line')
for i in range(len(X)):
    plt.plot([X[i], X[i]], [Y_actual[i], Y_pred[i]], 'gray', linestyle='--')

plt.title('Model Evaluation: RMSE & RÂ² Score')
plt.xlabel('Feature (X)')
plt.ylabel('Target (Y)')
plt.legend()

plt.text(0.5, max(Y_actual)+0.5, f'RMSE: {rmse_val:.2f}', bbox=dict(facecolor='lightblue', alpha=0.5))
plt.text(0.5, max(Y_actual)+1.0, f'RÂ² Score: {r2_val:.2f}', bbox=dict(facecolor='lightgreen', alpha=0.5))

plt.grid(True)
plt.show()
```

To - Do - 9 - Implementation in the Code:

```python
import numpy as np

def r2(Y, Y_pred):
    """
    This Function calculates the R Squared Error.

    Input Arguments:
    Y: Array of actual(Target) Dependent Variables.
    Y_pred: Array of predicted Dependent Variables.

    Output Arguments:
    r_squared: R Squared Error.
    """

    mean_y = np.mean(Y)
    ss_tot = np.sum((Y - mean_y)**2)          # Total sum of squares
    ss_res = np.sum((Y - Y_pred)**2)          # Residual sum of squares
    r2 = 1 - (ss_res / ss_tot)                # R-squared calculation

    return r2
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# ---- COST FUNCTION ----
def cost_function(X, Y, W):
    n = Y.shape[0]
    Y_pred = np.dot(W.T, X)
    return (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)


# ---- GRADIENT DESCENT ----
def gradient_descent(X, Y, W, alpha, iterations):
    n = Y.shape[0]
    cost_history = []

    for i in range(iterations):
        Y_pred = np.dot(W.T, X)
        gradient = (1 / n) * np.dot(X, (Y_pred - Y).T)
        W = W - alpha * gradient
        cost = (1 / (2 * n)) * np.sum((Y_pred - Y) ** 2)
        cost_history.append(cost)

    return W, cost_history


# ---- MAIN LOGIC ----
print("Starting program...")

data = pd.read_csv("/content/drive/MyDrive/Technologies_of_AI/student.csv")
print("Dataset loaded")

X = data[['Math', 'Reading']].values
Y = data['Writing'].values

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

W_init = np.zeros(X_train.shape[1])
alpha = 0.00001
iterations = 1000

W_final, cost_hist = gradient_descent(
    X_train.T, Y_train, W_init, alpha, iterations
)

Y_pred = np.dot(X_test, W_final)

print("Final Weights:", W_final)
print("Final Cost:", cost_hist[-1])
```
